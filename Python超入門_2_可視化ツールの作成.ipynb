## 可視化1

```
prefecture_list = df2["都道府県"]
target_col = "投票率（男）"

# 棒グラフを可視化
plot_data = go.Bar(y=prefecture_list, x=df2.loc[:, target_col], orientation='h')
fig = go.Figure(plot_data)

# グラフのタイトルと、グラフの大きさを指定
fig.update_layout(title="男性投票率", width=700, height=1000)
fig.show()
```

```
prefecture_list = df2["都道府県"]
target_col = "投票率（女）"

# 棒グラフを可視化
plot_data = go.Bar(y=prefecture_list, x=df2.loc[:, target_col], orientation='h')
fig = go.Figure(plot_data)

# グラフのタイトルと、グラフの大きさを指定
fig.update_layout(title="女性投票率", width=700, height=1000)
fig.show()
```

## 可視化2

```
prefecture_list = df2["都道府県"]
target_col1 = "前回投票率（男）"
target_col2 = "前回投票率（女）"

# 2つ分の棒グラフを作成
plot_data1 = go.Bar(name=target_col1, y=prefecture_list, x=df2.loc[:, target_col1], orientation='h')
plot_data2 = go.Bar(name=target_col2, y=prefecture_list, x=df2.loc[:, target_col2], orientation='h')
fig = go.Figure([plot_data1, plot_data2])

# タイトルとレイアウトを変更
fig.update_layout(title="男女 前回投票率", width=700, height=1000)
fig.show()
```

## 関数化

```
url = "https://www.soumu.go.jp/main_content/000776964.xls"
df = pd.read_excel(url)

df = df.iloc[5:-1, :]
df.columns = ["都道府県",
              "投票率（男）", "投票率（女）", "投票率（計）",
              "前回投票率（男）", "前回投票率（女）", "前回投票率（計）",
              "投票率変化（男）", "投票率変化（女）", "投票率変化（計）",
              ]

def plot_target_col(df, target_col, title):
    prefecture_list = df["都道府県"]
    plot_data = go.Bar(name=target_col, y=prefecture_list, x=df.loc[:, target_col], orientation='h')
    fig = go.Figure(plot_data)
    fig.update_layout(title=title, width=700, height=1000)
    fig.show()
```

```
plot_target_col(df, "投票率変化（男）", "男性の投票率変化のグラフ")
```

```
plot_target_col(df, "投票率変化（女）", "女性の投票率変化のグラフ")
```

### 複数の列を同時に出力する関数

```
def plot_target_col_list(df, target_col_list, title):
    prefecture_list = df["都道府県"]
    plot_data = [go.Bar(name=target_col, y=prefecture_list, x=df.loc[:, target_col], orientation='h') for target_col in target_col_list]
    fig = go.Figure(plot_data)
    fig.update_layout(title=title, width=700, height=1000)
    fig.show()
```

```
plot_target_col_list(df, ["投票率変化（女）", "投票率変化（男）"], "男女の投票率変化")
```
